```python
import os
import zipfile
import csv
import pandas as pd
```

```python
f = open("/content/Input","r")
a1 = f.read()
a= a1.lower()
print(a1)
f.close()
```

```python
print(a1[0])
```

```python
no = a1.find("45.")
print(no)
print(a1[no-10:no+200])
```

```python
listn = a1.split("\n")
print(listn)
```

```python
listq=[]
for i in listn:
  bol = i.find("?")
  if bol !=-1:
    listq.append(i)
print(len(listq))
print(listq)
```

```python
print(listq[44])
listq[44] = "45. Name and explain the three magic methods of Python that are used in the construction and initialization of custom\nObjects?"
print(listq[44])
```

```python
lista=[]
count =0
bol=0
for i in range(1,len(listq)):
  feend = listq[i]
  start = a1.find("Answer:",bol)
  print("start=",start)
  bol = a1.find(feend)
  if bol!=-1:
    count = count+1
    print("bol=",bol)
    string = a1[start:bol]
    print(string)
    lista.append(string)
string = a1[bol+len(feend)+2:]
lista.append(string)
count = count+1
print(count)  
print(lista)
```

```python
df = pd.read_csv(r'/content/Output.csv')
print(df.to_string())
```

```python
with open('submission.csv','w',newline='') as file:
  writer = csv.writer(file)
  writer.writerow(["Question","Answer"])
  for i in range(0,len(listq)):
    writer.writerow([listq[i],lista[i]])
print(file)
```
